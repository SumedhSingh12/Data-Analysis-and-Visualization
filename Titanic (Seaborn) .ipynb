## Data Visualization on the famous Titanic dataset using Seaborn.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('whitegrid')
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

### Creating a Distribution Plot (Joint Plot) for the ticket fare vs the age of the traveller

```python
sns.jointplot(x='fare', y='age',data=titanic)
```

### Creating a distribution plot for plotting the ticket fare . The KDE attribute was set to false.

```python
sns.distplot(titanic['fare'], kde=False, color='red', bins=30)
```

### The box plot represents the age of people in different classes of the Titanic 

```python
sns.boxplot(x='class',y='age',data=titanic)
```

### A swarm plot for the same purpose.

```python
sns.swarmplot(x='class', y='age', data=titanic)
```

### A Count plot to represent the number of males and females on the ship

```python
sns.countplot(x='sex',data=titanic)
```

### A heatmap to show the dataset correlation

```python
plt.figure(figsize=(16,10))
sns.heatmap(titanic.corr(), cmap='magma', annot=True, linewidth = 3, linecolor='grey')
plt.title('titanic.correlation')
```

### A FacetGrid to show the age distribution of males and females.

```python
t = sns.FacetGrid(titanic, col='sex')
t.map(sns.distplot,'age',kde=False, bins=10, color='green')
```
